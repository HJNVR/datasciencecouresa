```python
import requests
url = "https://www.theweathernetwork.com/ca/weather/ontario/mississauga"
r = requests.get(url)
```

```python
# check whether request good or not
r.status_code # 200 means good
```

```python
from bs4 import BeautifulSoup
page = r.text
# Creates a BeautifulSoup object
soup = BeautifulSoup(page, "html.parser")
```
